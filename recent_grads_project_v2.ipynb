## Pandas, scatter plots

```python
recent_grads.plot(x='Sample_size', y='Employed', kind='scatter', title='Employed vs. Sample_size')
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Sample_size vs Unemployment_rate')
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Full_time vs Median')
```

```python
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter', title='ShareWomen vs Median')
```

```python
recent_grads.plot(x='ShareWomen', y = 'Unemployment_rate', kind='scatter', title='ShareWomen vs. Unemployment')
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Men vs Median')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Women vs. Median')
```

### From the above there seem to be no strong correlation between the data, except for the sample size and employed numbers

## Pandas, histograms

```python
recent_grads['Sample_size'].hist(bins=20, range=(0,3000))
```

### Most of the 'Sample_size' is less than 500, so may not be representative

```python
recent_grads['Median'].hist(bins=25, range=(0,80000))
```

### The median salary is between $30-40,000

```python
recent_grads['Employed'].hist(bins=10, range=(0,200000))
```

```python
recent_grads['Full_time'].hist(bins=20, range=(0,200000))
```

```python
recent_grads['ShareWomen'].hist(bins=25, range=(0,1))
```

```python
recent_grads['Unemployment_rate'].hist(bins=25, range=(0,0.2))
```

### The most common 'Unemployment_rate' is between 5-7% . Let's look which majors are in this range

```python
unemploy = recent_grads['Unemployment_rate'].between(0.05, 0.07)
```

```python
recent_grads.loc[unemploy].sort_values(['Sample_size'], ascending=False).head(10)
```

### Interestingly the majors with the most unemployed graduates include in-demand skills such as Computer Science & Business. Perhaps too many people are doing these degrees?
