## Pandas, scatter plots

```python
recent_grads.plot(x='Sample_size', y='Employed', kind='scatter', title='Employed vs. Sample_size')
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Sample_size vs Unemployment_rate')
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Full_time vs Median')
```

```python
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter', title='ShareWomen vs Median')
```

```python
recent_grads.plot(x='ShareWomen', y = 'Unemployment_rate', kind='scatter', title='ShareWomen vs. Unemployment')
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Men vs Median')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Women vs. Median')
```

### From the above graphs there looks to be a strong correlation in the first graph - between number of people Employed and the size of the Sample

Use the plots to explore the following questions:

Do students in more popular majors make more money?
No there is not a correlation between these in the Median salary vs Sample size

Do students that majored in subjects that were majority female make more money?
There is no correlation shown between the share of women in a major and salary level

Is there any link between the number of full-time employees and median salary?
There is no correlation show between number of full time employees and median salary
